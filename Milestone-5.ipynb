# Milestone 5 – Probability
This notebook contains sample code for each section of Milestone 5.

## Section A: Setup & Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Global_Tech_Gadget_Consumption.csv')
df.head()
```

## Section B: Treat Dataset as Sample Space

```python
N = len(df)
print('Total observations:', N)
```

## Section C: Task 1 – Defining Events

```python
A = df[df['Laptop Shipments (Millions)'] > 70]
B = df[df['Smartwatch Penetration (%)'] == 'A']
C = df[(df['Laptop Shipments (Millions)'] >= 50) & (df['Laptop Shipments (Millions)'] <= 80)]

print('Event A size:', len(A))
print('Event B size:', len(B))
print('Event C size:', len(C))
```

## Section D: Task 2 – Calculating Basic Probability

```python
P_A = len(A) / N
P_B = len(B) / N
P_C = len(C) / N

print('P(A) =', P_A)
print('P(B) =', P_B)
print('P(C) =', P_C)
```

## Section E: Task 3 – Combined Events

```python
A_int_B = df[(df['Laptop Shipments (Millions)'] > 70) & (df['Smartwatch Penetration (%)'] == 'A')]
P_A_int_B = len(A_int_B) / N
print('P(A ∩ B) =', P_A_int_B)

A_union_B = df[(df['Laptop Shipments (Millions)'] > 70) | (df['Smartwatch Penetration (%)'] == 'A')]
P_A_union_B = len(A_union_B) / N
print('P(A ∪ B) =', P_A_union_B)

A_comp = df[df['Laptop Shipments (Millions)'] <= 70]
P_A_comp = len(A_comp) / N
print('P(Aᶜ) =', P_A_comp)

rule_value = P_A + P_B - P_A_int_B
print('Rule Verification:', rule_value)
print('Actual P(A ∪ B):', P_A_union_B)
```

## Section F: Visualization

```python
counts = [len(A), len(A_comp)]
labels = ['A: Laptop Shipments (Millions) > 70', 'Aᶜ: Laptop Shipments (Millions) ≤ 70']

plt.bar(labels, counts)
plt.title('Event A and Complement Frequency')
plt.ylabel('Count')
plt.show()
```
