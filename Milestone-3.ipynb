## Part 1: Data Loading and Preparation


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('Global_Tech_Gadget_Consumption.csv')

df.head()
```

## Part 2: Frequency Distribution Table


### Column Selection

**Justification:**

The reason for choosing the 'Average Customer Spending on Gadgets ($)' column is because it is a key demographic variable and understanding its distribution can provide insights into the gadget expense demographics of the subjects in this dataset.

```python
column_to_analyze = 'Gaming Console Adoption (%)'

num_bins = int(np.ceil(1 + 3.322 * np.log(df[column_to_analyze].count())))
bins = pd.cut(df[column_to_analyze], bins=num_bins)

#To Calculate Frequency (f)
freq_table = pd.DataFrame(df[column_to_analyze].value_counts()).reset_index()
freq_table.columns = ['Class/Category', 'Frequency (f)']
freq_table = freq_table.sort_values(by='Class/Category').reset_index(drop=True)

# To Calculate Relative Frequency (rf)
total_count = freq_table['Frequency (f)'].sum()
freq_table['Relative Frequency (rf)'] = freq_table['Frequency (f)'] / total_count

# To Calculate Cumulative Frequency (cf)
freq_table['Cumulative Frequency (cf)'] = freq_table['Frequency (f)'].cumsum()

# To Calculate Relative Cumulative Frequency (rcf)
freq_table['Relative Cumulative Frequency (rcf)'] = freq_table['Relative Frequency (rf)'].cumsum()


# For Displaying the Frequency Table
print("Frequency Distribution Table for '" + column_to_analyze + "'")
freq_table
```

## Part 3: Graphical Representation


### 3.1 Bar Chart / Histogram

```python
plt.figure(figsize=(20, 8))

sns.countplot(data=df, x=column_to_analyze, order=freq_table['Class/Category'], palette='viridis')

plt.title(f'Bar Chart of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Frequency', fontsize=0.01)
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

### 3.2 Line Chart / Frequency Polygon

```python
plt.figure(figsize=(12, 6))

plt.plot(freq_table['Class/Category'], freq_table['Frequency (f)'], marker='o', linestyle='-')

plt.title(f'Frequency Polygon of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Frequency', fontsize=0.01)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 3.3 Ogive Chart (Cumulative Frequency Graph)

```python
plt.figure(figsize=(12, 6))

plt.plot(freq_table['Class/Category'], freq_table['Cumulative Frequency (cf)'], marker='o', linestyle='-')

plt.title(f'Ogive Chart (Less Than) of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Cumulative Frequency', fontsize=0.01)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Part 4: Analysis and Conclusion

---

The analysis of the 'Average Customer Spending on Gadgets ($)' column shows that all the catagories have the same bar sizes except 4 bars: 16.48%,19.64%,28.46 and 35.63%.The bar chart reveals a distribution that is multimodal.The Ogive chart indicates that approximately 50% of the data falls below the value of 19.75%.
